```python
import matplotlib.pyplot as plt
from shapely.geometry import Polygon
import numpy as np
from scipy.spatial import Delaunay
```

```python
figsizes = (8,6)

coordinates = [
    (13.363376884873674, 52.52487384662575),
    (13.366370231681032, 52.52244551129306),
    (13.371166021399262, 52.5232680273848),
    (13.371037275366263, 52.52510231335614),
    (13.370393545203285, 52.527608829404016),
    (13.363376884873674, 52.52487384662575)  
]

polygon = Polygon(coordinates)
x, y = polygon.exterior.xy
```

```python
x, y
```

```python
polygon
```

```python
plt.figure(figsize=figsizes)
plt.plot(x, y, marker='o', color='blue', label='Polygon')
plt.fill(x, y, color='lightblue')
plt.title("Polygon Representation")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid(True)
plt.legend()
plt.show()
```

```python
points = np.array(coordinates)
tri = Delaunay(coordinates)

plt.figure(figsize=figsizes)
plt.triplot(points[:, 0], points[:, 1], tri.simplices, color='blue', label='Delaunay Triangles')
plt.scatter(points[:, 0], points[:, 1], color='red', marker='o', label='Points')
plt.title("Delaunay Triangulation")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid(True)
plt.legend()
plt.show()
```
